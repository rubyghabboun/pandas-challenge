```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete_df.head()
```

## District Summary

```python
# Calculate the total number of unique schools
unique_schools = school_data_complete_df["school_name"].unique()
school_count = len(unique_schools)
school_count
```

```python
# Calculate the total number of students
student_count = school_data_complete_df["student_name"].count()
student_count
```

```python
# Calculate the total budget
total_budget =  school_data_complete_df["budget"].sum()
total_budget
```

```python
# Calculate the average (mean) math score
average_math_score =  school_data_complete_df["math_score"].mean()
average_math_score
```

```python
# Calculate the average (mean) reading score
average_reading_score =  school_data_complete_df["reading_score"].mean()
average_reading_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete_df[(school_data_complete_df["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = school_data_complete_df[(school_data_complete_df["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage                                                                      
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete_df[
    (school_data_complete_df["math_score"] >= 70) & (school_data_complete_df["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = {
    'number of schools': [school_count],
    'total students': [student_count],
    'total budget': [total_budget],
    'average math score': [average_math_score],
    'average reading score': [average_reading_score],
    '% math passing': [passing_math_percentage],
    '% reading passing': [passing_reading_percentage],
    'overall passing rate': [overall_passing_rate]}

# Create a DataFrame using the metrics data
district_summary = pd.DataFrame(district_summary)

# Formatting
district_summary['total students'] = district_summary['total students'].map("{:,}".format)
district_summary['total budget'] = district_summary['total budget'].map("${:,.2f}".format)

# Display the DataFrame
district_summary
```

## School Summary

```python
# Use the code provided to select all of the school types
school_types = school_data_complete_df["type"].unique()
school_types
```

```python
per_school_counts = school_data_complete_df.groupby(['school_name', 'type']).size().reset_index(name='Total Students')

per_school_counts.rename(columns={'Student ID': 'Total Students'}, inplace=True)


per_school_counts
```

```python
# Calculate the total school budget and per capita spending per school
per_school_budget = school_data_complete_df.groupby('school_name')['budget'].first()

# Calculate the total number of students per school
per_school_students = school_data_complete_df['school_name'].value_counts()

# Calculate the per student budget
per_student_budget = per_school_budget / per_school_students

# Create a DataFrame with the per school and per student budgets
budget_summary = pd.DataFrame({
   'Per School Budget': per_school_budget,
    'Per Student Budget': per_student_budget
})

# Reset the index for clarity (optional)
budget_summary.reset_index(inplace=True)
budget_summary.rename(columns={'index': 'School Name'}, inplace=True)

# Display the budget summary
budget_summary


```

```python
average_math_score_per_school = school_data_complete_df.groupby('school_name')['math_score'].mean().reset_index()
average_reading_score_per_school = school_data_complete_df.groupby('school_name')['reading_score'].mean().reset_index()


print(average_math_score_per_school)
print(average_reading_score_per_school )
```

```python


passing_math_df = school_data_complete_df[school_data_complete_df['math_score'] >= 70]
passing_math_per_school = passing_math_df.groupby('school_name')['Student ID'].count().reset_index()
passing_math_per_school.rename(columns={'Student ID': 'Math Pass Count'}, inplace=True)
# Display the result

passing_math_per_school
```

```python
# Calculate the number of students per school with reading scores of 70 or higher
passing_reading_df = school_data_complete_df[school_data_complete_df['reading_score'] >= 70]
passing_reading_per_school = passing_reading_df.groupby('school_name')['Student ID'].count().reset_index()

# Rename the column to 'Number of Students with Math Scores >= 70' for clarity
passing_reading_per_school.rename(columns={'Student ID': 'reading pass count'}, inplace=True)

# Display the result
passing_reading_per_school
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher

passing_both_per_school = passing_math_per_school.merge(passing_reading_per_school, on='school_name')

# Calculate the number of students who passed both math and reading
passing_both_per_school['Both Pass Count'] = passing_both_per_school['Math Pass Count'] + passing_both_per_school['reading pass count']

passing_both_per_school
```

```python



# Calculate the overall passing rate
per_school_passing_math = (passing_math_per_school['Math Pass Count'] / per_school_counts) * 100
per_school_passing_reading = (passing_reading_per_school['reading pass count'] / per_school_counts) * 100
overall_passing_rate = (passing_both_per_school['Both Pass Count'] / per_school_counts) * 100
```

```python

# Create the per_school_summary DataFrame
per_school_summary = pd.DataFrame({
    'school Name':per_school_counts['school_name'],
    'School Type': per_school_counts['type'],  
    'Total Students': per_school_counts['Total Students'],  
    'Total School Budget': budget_summary['Per School Budget'],
    'Per Student Budget': budget_summary['Per Student Budget'],
    'Average Math Score': average_math_score_per_school['math_score'],
    'Average Reading Score': average_reading_score_per_school['reading_score']
})
# Calculate and add the passing rates
per_school_summary['% Passing Math'] = (passing_math_per_school['Math Pass Count'] / per_school_counts['Total Students']) * 100
per_school_summary['% Passing Reading'] = (passing_reading_per_school['reading pass count'] / per_school_counts['Total Students']) * 100
per_school_summary['% Overall Passing Rate'] = (passing_both_per_school['Both Pass Count'] / per_school_counts['Total Students']) * 100

per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)




# Display the DataFrame
per_school_summary
```

## Scores by School Spending

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels1 = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary.copy()
```

```python
school_spending_df["Total School Budget"] = pd.to_numeric(school_spending_df["Total School Budget"].str.replace('[\$,]', '', regex=True))
school_spending_df["Per Student Budget"] = pd.to_numeric(school_spending_df["Per Student Budget"].str.replace('[\$,]', '', regex=True))
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["Per Student Budget"], bins=spending_bins, labels=labels1, right=False)

# Display the DataFrame
school_spending_df.head()
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing Rate"].mean()
```

```python
# Assemble into DataFrame
spending_summary = pd.DataFrame({
    "Average Math Score": spending_math_scores,
    "Average Reading Score": spending_reading_scores,
    "% Passing Math": spending_passing_math,
    "% Passing Reading": spending_passing_reading,
    "% Overall Passing Rate": overall_passing_spending
})

# Display the spending summary
spending_summary
```

## Scores by School Size

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.
per_school_summary["School Size"] = pd.cut(per_school_summary['Total Students'], bins=size_bins, labels=labels)

# Display the updated DataFrame
per_school_summary
```

```python
# Calculate averages for the desired columns. 
size_math_scores = per_school_summary.groupby(["School Size"])["Average Math Score"].mean()
size_reading_scores = per_school_summary.groupby(["School Size"])["Average Reading Score"].mean()
size_passing_math = per_school_summary.groupby(["School Size"])["% Passing Math"].mean()
size_passing_reading = per_school_summary.groupby(["School Size"])["% Passing Reading"].mean()
size_overall_passing = per_school_summary.groupby(["School Size"])["% Overall Passing Rate"].mean()
```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`
size_summary = pd.DataFrame({
    'Average Math Score': size_math_scores,
    'Average Reading Score': size_reading_scores,
    '% Passing Math': size_passing_math,
    '% Passing Reading': size_passing_reading,
    '% Overall Passing Rate': size_overall_passing
})

# Display results
size_summary
```
